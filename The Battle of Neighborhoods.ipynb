#### Introduction
This project is to help client planning to open a restaurant in Vancouver, BC. But the location of restaurant and type of food served are unknown. Certain steps we are going to walk through to reach a final conclusion, such as retrieving data, analyzing, results, recommendations. Matrix would be helpful to compare types of restaurant and neighbourhood. Also, average rental fee will be considered, helps client to control the cost and maximize the return.

#### Source of Data
- Geonames  
Data https://www.geonames.org/postal-codes/CA/BC/british-columbia.html will be scrapped from geonames website for each city in Greater Vancouver Area. This data contains city name, postal code, and coordinator of city, which could help connecting to Foursquare.  ![Geoname](https://github.com/HaoZ124/Coursera_Capstone/blob/main/Capture.PNG?raw=true)
- Population  
The population of Greater Vancouver Area https://www.citypopulation.de/en/canada/britishcolumbia/admin/5915__greater_vancouver/ created by statistics Canada would help us find out population distribution. Data will be processed and refined to merge into previous geonames data frame.  ![Population](https://github.com/HaoZ124/Coursera_Capstone/blob/main/Capture1.PNG?raw=true)
- Foursqure API  
Through Foursqure API data we can all manner of venues and events within an area of interest. Since Foursqure as a powerful tool is able to give us all functions we need, we decide only use Foursqure API data as map information search. To receive venues names, location, and menus we have to send request with neighborhood's coordinator and searching radius 300 meters.
